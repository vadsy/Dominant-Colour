```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
image = image.reshape((image.shape[0] * image.shape[1], 3))
print image
print image.shape
```

```python
from sklearn.cluster import KMeans
K = 3
clf = KMeans(n_clusters=K)
clf.fit(image)
```

```python
#initialising the label dictionary
label = [0, 0, 0]

#finding count of data in each cluster
for x in clf.labels_:
    label[x] += 1

#Total Number of data points
label_sum = image.shape[0]

#converting to percentage
label = [float(x) / label_sum for x in label]
print label
```

```python
import webcolors

centroids = [x.astype("uint8").tolist() for x in clf.cluster_centers_]

clr = [webcolors.rgb_to_hex(c) for c in centroids]

plt.pie(label, colors=clr, shadow=True) 
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
domc=pd.DataFrame({
    'Image name': [],
    'Dominant colour percentage': [],
    'Color': []
})
```

```python
data = []
for i in range(1,11):
    image = ('image'+str(i)+'.jpg')
    print image
    image = cv2.imread(image)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    print 'Image shape: '+str(image.shape)
    # converting 3d image array to 1d
    image = image.reshape((image.shape[0]*image.shape[1],3))
    # print image
    print 'Shape of image after image reshape: '+str(image.shape)

    # declaring object of KMeans
    kmeans = KMeans(n_clusters=3, random_state=0).fit(image)

    # initialising the label dictionary
    label = [0,0,0]

    # finding count of data in each cluster
    for x in kmeans.labels_:
        label[x] += 1

    # total number of data points
    label_sum = image.shape[0]

    #converting to percentage
    label = [round((float(x)/label_sum)*100, 2) for x in label]
    m = max(label)
    y = label.index(m)
    print 'Percentage of different colours: '+str(label)

    # print kmeans.cluster_centers_

    # color values of each centeroid pixels of different clusters
    centroids = [x.astype('uint8').tolist() for x in kmeans.cluster_centers_]
    print centroids
    
    # covert pixel RGB color intensity values to hex code
    clr = [webcolors.rgb_to_hex(c) for c in centroids]
    print 'colours hex codes: '+ str(clr)
    dc = clr[y]

    # ploting pie chart
    plt.pie(label, colors=clr, shadow=True, autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()
    
    # appending data to list
    l = []
    l.append('Image'+str(i))
    l.append(m)
    l.append(dc)
    
    # appending list all the image details to a list
    data.append(l)
    
    # appending data to dataframe
    domc = domc.append(pd.DataFrame({
        'Image name' : [data[i-1][0]],
        'Dominant colour percentage' : [data[i-1][1]],
        'Color' : [data[i-1][2]]
    }))
    
```
